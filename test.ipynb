```python
import torch
import matplotlib.pyplot as plt

# Example tensor
data = torch.randn(17, 69696, 1)  # Simulating your tensor




# heatmap to show all feature in one time step 
# Select a specific time step
time_step_idx = 0
features = data[time_step_idx, :, 0].numpy()  # Extract all features at the selected time step

# Plot a histogram
plt.figure(figsize=(12, 6))
plt.hist(features, bins=50, color='blue', alpha=0.7, edgecolor='black')
plt.xlabel("Feature Values")
plt.ylabel("Frequency")
plt.title(f"Feature Value Distribution at Time Step {time_step_idx}")
plt.grid(True)
plt.show()
```
